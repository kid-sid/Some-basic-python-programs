```
# Reverse a string

def reverse_the_string(string):
    reversed_string = ''           # Creating an empty string
    for char in string[::-1]:      # Accessing the characters from the last to the first
        reversed_string += char    # Adding characters to empty string
    return reversed_string

print(reverse_the_string(input()))
        
```

```
# First factorial

def first_factorial(num):
    if num == 0:
        return 1
    else:
        return (num) * (num-1)

print(first_factorial(int(input('Enter a number:'))))
```

```
# Without using regex, checking longest word in a sentence

def longestWord(sentence):
    longest_word = ''
    for char in "~!@#$%^&*()-_=+[{]}\|';:/?.>,<":
        no_punct_sentence = sentence.replace(char," ")
    split_sentence = no_punct_sentence.split()

    for word in split_sentence:
        if len(word) > len(longest_word):
            longest_word = word
    return longest_word

sentence_with_punct = input('Enter the sentence :')
longestWord(sentence_with_punct)
```

```
# Finding second lowest and second highest

def SecondGreatLow(arr):
    multiples_removed_list = list(set(arr))
    sorted_list = sorted(multiples_removed_list)

    second_greatest = sorted_list[-2]
    second_lowest = sorted_list[1]

    return "%d %d" % (second_lowest, second_greatest)

print(SecondGreatLow([7, 7, 12, 98, 106]))
```

```
# Counting number of vowels in a sentence

def VowelCount(a_string):
    vowels = 'aeiou'
    vowel_count = 0
    for char in a_string.lower():
        if char in vowels:
            vowel_count += 1
    return vowel_count

VowelCount('Hey how are you doing?')
```

```
# Checking if a given sentence is palindrome or not

def Palindrome(a_string):
    lowered = a_string.lower()
    string_no_spaces = lowered.replace(" ", "")
    if string_no_spaces == string_no_spaces[::-1]:
        return "true"
    return "false"

x = input('Enter a string: ')
Palindrome(x)
```

```
# Find twin prime between 1 to 1000
# 2 continuos prime numbers are known as twin prime. e.g: (3,5),(5,7),(7,11) and so on
# https://www.geeksforgeeks.org/twin-prime-numbers-between-1-and-n/

def twinprime():
    for num in range (2,1000):
        for i in range (2,num):
            if num%i == 0:
                break
        else:
            if num%2 != 0:
                print(num)
print('Twin prime between 1 to 1000 are : ')
twinprime()
```
